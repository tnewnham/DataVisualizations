```python
import pandas as pd
```

```python
df = pd.read_csv('./school grants 2010.csv')
```

```python
df.info()
```

```python
import seaborn as sns
import matplotlib as plt
```

```python
sns.barplot(data=df,
         y='Award_Amount',
         x='Model Selected',
         hue="Region")
```

```python
sns.lvplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='Paired',
         hue='Region')
```

```python
sns.swarmplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         hue ='Region')
```

```python
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         hue ='Region')
```

```python
# sns.set(color_codes= True)
sns.violinplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         palette='husl')
```

```python
sns.stripplot(data=df,
         x='Award_Amount',
         y='Model Selected',
         jitter= True)
```
